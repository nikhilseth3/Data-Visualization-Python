## Matplotlib 

```python
import matplotlib.pyplot as plt 
```

#### Using the follwoing - the visualizations will be in the notebook

```python
%matplotlib inline      
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method

plt.plot(x,y)     # ploting x vs y
plt.show()
```

```python
# Functional method

plt.plot(x,y,'g')     # ploting x vs y
plt.xlabel('X Label')    # x axis label    
plt.ylabel('Y Label')    # y axis label
plt.title('LOL')    # title
```

#### Multiple Plots on the same canvas

```python
plt.subplot(1,2,1)  # subplot(no of rows, no of columns, plot number)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### OOP

```python
#OO

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])  # left,bottom,width,height 
# values must be between 0 and 1 for better visualization

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# manually doing it

fig1 = plt.figure()

axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig1.add_axes([0.2,0.5,0.3,0.2])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
# basically an axes manager
fig1,axes = plt.subplots(nrows=3,ncols=2) # doing the axes plot automatically

#axes.plot(x,y)
plt.tight_layout()          # tight layout
```

```python
axes
```

```python
fig1,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'r--')
axes[0].set_title("First Plot")
axes[0].set_xlabel("X Axis")
axes[0].set_ylabel("Y Axis")
axes[0].set_ylim([0,30])       # setting ylim. can do xlim also

axes[1].plot(y,x,'c--')
axes[1].set_title("Second Plot")
```

#### Figure Size and DPI 

```python
fig = plt.figure(figsize=(3,2)) # figsize=(width,height)
                 
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,3))    
                 
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig5,axes = plt.subplots(nrows=2,ncols=1) 
                 
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig5,axes = plt.subplots(nrows=2,ncols=1,figsize=(3,5)) 
                 
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

####  Legends

```python
f = plt.figure() 
                 
ax = f.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
f = plt.figure() 
                 
ax = f.add_axes([0,0,1,1])
ax.plot(x,x**2,'r',label = "X Squared")        # adding label for a legend
ax.plot(x,x**3,label = "X Cubed")

ax.legend(loc='best')   # checking for the label parameter
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,'r',linewidth = 2,marker='*',markersize=10,markerfacecolor='b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,'r',linewidth = 2,marker='*',
          markersize=10,markerfacecolor='b',
          markeredgewidth=3,
         markeredgecolor='green')  #more options in the notebook section
```
